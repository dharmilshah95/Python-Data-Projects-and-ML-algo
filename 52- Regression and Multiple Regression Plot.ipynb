```python
#Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# to see the plots in python notebook
%matplotlib inline
```

```python
# inbuild dataset tips. Directly importing here
tips=sns.load_dataset('tips')
```

```python
tips.head(4)
```

```python
# lmplot for plotting regression
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,scatter_kws={'marker':'o','color':'orange'},line_kws={'linewidth':1,'color':'blue'})
```

```python
# type of order fit we want eg 2/3/4 etc so we will just add order=2/3/4 etc
sns.lmplot('total_bill','tip',tips,order=6, scatter_kws={'marker':'o','color':'orange'},line_kws={'linewidth':1,'color':'blue'})
```

```python
# do not plot regression only plot scatter then fit_reg=False
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips.head()
```

```python
tips['tip_prcnt']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_prcnt',tips)
```

```python
sns.lmplot('size','tip_prcnt',tips,x_jitter=.1)
```

```python
# Assign Hue=sex aitomatically attach sex column and writing markers will display markers in plot
sns.lmplot('size','tip_prcnt',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_prcnt',tips,hue='day')
```

```python
# regplot is lower level plot function then lmplot. After that lmplot was made
sns.regplot('total_bill','tip_prcnt',tips)
```

```python
fig,(axis1,axis2)=plt.subplots(1,2,sharey=True)

sns.regplot('total_bill','tip_prcnt',tips,ax=axis1)
sns.violinplot(tips['tip_prcnt'],tips['size'],color='pink',ax=axis2)
```
